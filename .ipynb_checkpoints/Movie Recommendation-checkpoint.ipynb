```python
import pandas as pd
metadata = pd.read_csv('./data/IMDB-Movie-Data.csv')
metadata
```

```python
C = metadata['Rating'].mean()
print(C)
```

```python
m = metadata['Votes'].quantile(0.9)
print(m)
```

```python
q_movies = metadata.copy().loc[metadata['Votes'] >= m]
print(q_movies.shape)
q_movies.sort_values(by='Rating', ascending=False)
```

```python
def WR(x, m=m, C=C):
    v = x['Votes']
    R = x['Rating']
    return (v/(v+m)*R + (m/(m+v)*C))

metadata['score'] = metadata.apply(WR, axis=1)

col = ['Title', 'Rating', 'Votes', 'score']
rs_original = metadata.sort_values('Rating', ascending=False)[col]
rs_new = metadata.sort_values('score', ascending=False)[col]
```

# TF-IDF (Term Frequency - Inverse Document Frequency)
 - TF: 특정 단어가 얼마나 많이 등장하는지

하지만 그 특정 단어가 다른 문서에서 자주 등장한다면 단어의 중요도는 낮아진다 - 이것이 DF

 - IDF: DF의 역수

TF-IDF 는 TF와 IDF의 곱이다. 이 점수가 높은 단어일 수록 다른 문서에서는 많지 않고 해당 문서에서 자주 등장하는 단어이다

```python
from sklearn.feature_extraction.text import TfidfVectorizer
tfidf = TfidfVectorizer(stop_words='english')
metadata.Description = metadata.Description.fillna('') # null이 있으면 오류가 나기 때문에 하는 작업
tfidf_matrix = tfidf.fit_transform(metadata.Description)
tfidf_matrix.shape
```

```python
from sklearn.metrics.pairwise import linear_kernel
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
cosine_sim
```

```python
indicies = pd.Series(metadata.index, index=metadata['Title']).drop_duplicates()
```

```python
def rs(title, n = 10, sim = cosine_sim) : 
    idx = indicies[title]
    sim_score = list(enumerate(cosine_sim[idx]))
    sim_score = sorted(sim_score, key = lambda x : x[1], reverse = True)
    sim_score = sim_score[1:(n+1)]
    movie_index = [i[0] for i in sim_score]
    return metadata.Title.iloc[movie_index]
```

```python
rs("The Dark Knight")
```

```python
rs("Suicide Squad", n=5)
```
