```python
from sklearn.datasets import make_regression
import numpy as np
#회귀분석용 가상 데이터 생성 함수: make_regression
#n_samples : 표본의 개수
#n_features : 독립변수의 개수
#bias : y 절편
#noise : 종속 변수에 더해지는 잡음의 표준편차
#coef : True 이면 선형 모형의 계수도 출력한다
#random_state : 난수 발생용 시드값
X0, y, coef = make_regression(n_samples = 100, n_features = 2, bias = 100, noise = 10, coef = True, random_state = 1)
```

```python
X0
```

```python
y
```

```python
X = np.hstack([np.ones((X0.shape[0], 1)), X0]) # 1 로 채워진 벡터들을 X0 앞에 넣는다.
#np.hstack : 배열을 왼쪽에서 오른쪽으로 붙이기(좌우)
X[:5]
```

```python
import statsmodels.api as sm
X = sm.add_constant(X0)
X[:5]
```

## OLS

```python
X0, y, w = make_regression(n_samples= 200, n_features = 1, bias = 100, noise = 10, coef = True, random_state = 1)
X = sm.add_constant(X0)
print(X)
print(y)
y = y.reshape(len(y), 1) # 첫 번째 열만 잘라옴.
y
```

```python
w
```

```python
print(X.shape)
w1 = X.T @ X # @ : matrix multiplication
w2 = np.linalg.inv(w1)
w3 = w2 @ X.T
w4 = w3 @ y
print(w1)
print(w2)
#print(w3)
print(w4)
```

```python
## 위 과정을 다음과 같이 한 번에 쓸 수 있다.
w = np.linalg.lstsq(X, y, rcond=None)[0]
w
```

```python
import matplotlib.pylab as plt
%matplotlib inline

x_new = np.linspace(np.min(X0), np.max(X0), 100)
X_new = sm.add_constant(x_new)
y_new = new = np.dot(X_new, w)
print(x_new[:10])
print(X_new[:10])
print(y_new[:10])

plt.figure(figsize=(10,10))
plt.scatter(X0, y, label="data")
plt.plot(x_new, y_new, 'r-', label = "result of linear regression")
plt.xlabel("x")
plt.ylabel("y")
plt.title("example of linear regression")
plt.legend()
plt.show()
```

## 패키지를 사용한 linear regression

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression(fit_intercept = True) # fit_intercept : 상수항이 있는지 없는지. 있는 것이 계산하기 좋다
```

```python
model = model.fit(X, y)
print(model.coef_, model.intercept_)
```

```python
y_new = model.predict(X_new)
print(X_new.shape)
print(y_new.shape)

y_new[:10]
```

```python
from sklearn.datasets import load_boston
boston = load_boston()
print(boston.DESCR)
```

```python
import pandas as pd
dfX = pd.DataFrame(boston.data , columns = boston.feature_names)
print(dfX)
dfy = pd.DataFrame(boston.target, columns = ["MEDV"])
print(dfy)
df = pd.concat([dfX, dfy], axis = 1)
df.tail()
```

### 데이터프레임을 만든 후에는 데이터에 대한 사전 조사를 해야한다
 - 데이터에 누락된 값이 있는지 확인
 - 각 데이터가 연속적인 실수값인지 범주형 값인지 확인(범주형 값: 1 또는 0과 같은 것)
 - 실수형 데이터의 분포가 정규분포인지 확인(통계학적 기법을 사용할 수 있는지에 대한 판단 기준)
 - 실수형 데이터에 양수 혹은 범위 등 제한 조건이 있는지
 - 범주형 데이터의 경우 범주의 값이 어떤 값 혹은 숫자로 표현되어 있는지
 - 데이터 간의 상관관계 확인
 - 데이터에 이상한 값(outlier)들이 있는지 확인

```python
import seaborn as sns
cols = ["MEDV", "RM", "AGE", "RAD"]
sns.pairplot(df[cols])
plt.show()
```

```python
model_boston = LinearRegression().fit(boston.data , boston.target)
model_boston.coef_
```

```python
boston.feature_names
```

```python
predictions = model_boston.predict(boston.data)
plt.scatter(boston.target, predictions)
plt.show()
```

```python
from sklearn.metrics import explained_variance_score, mean_squared_error, mean_absolute_error, r2_score
print('explained_variance_score: {}'.format(explained_variance_score(boston.target, predictions)))
print('mean_squared_errors: {}'.format(mean_squared_error(boston.target, predictions)))
print('r2_score: {}'.format(r2_score(boston.target, predictions)))
```
